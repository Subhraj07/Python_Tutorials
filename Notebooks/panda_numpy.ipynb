### Map, filter, Reduce

```python
def add_one(this_item):
    return this_item + 1
```

```python
numbers = [1,2,3,4,5]
res = map(add_one, numbers)
```

```python
for i in res:
    print(i)
```

```python
def is_even(this_item):
    return this_item % 2 == 0
```

```python
res = filter(is_even,numbers)
```

```python
for i in res:
    print(i)
```

```python
def add_item(first, second):
    return first + second
```

```python
from functools import reduce
reduce(add_item, numbers)
```
